```python
# Extract the clean text
import re
CLEANR = re.compile('<.*?>') #match and capture any HTML tags, including their attributes, in a non-greedy way. 
    #The .*? is a non-greedy quantifier that matches as few characters as possible.
for c in content:
    # print(c)
    clean_text = re.sub(CLEANR, '', c)  #replace all matches of the CLEANR pattern with an empty string ('') in the current string c. 
    # This effectively removes all HTML tags from the current string and stores the result in the clean_text variable.
    # print(clean_text)
    print(clean_text.split('>')[1])
    # Not perfectly clean
```

```python
# Travel through many nested URLs
import time

# Define a function to get all the links from a page
def crawler(url):
    html_page = requests.get(url)
    time.sleep(2)
    page = html_page.text
    links = page.split('<a href="')[1:]
    url_set = []
    ROOT = 'https://www.lanacion.com.ar'
    for link in links:
        if link.startswith('/'):
            link = ROOT + link
        if ROOT in link:
            url_set.append(link.split('"')[0])
    return(url_set)
```

```python
# Test it
url = 'https://www.lanacion.com.ar/politica/rebaten-la-acusacion-de-milei-y-niegan-que-patricia-bullrich-haya-puesto-bombas-en-jardines-de-nid03102023/'
urls = crawler(url)
print(urls)
```
